# Summary
The Iris dataset is one of the most famous datasets in machine learning, commonly used for classification problems.
1. Total samples: 150 flowers
2. Total features: 4 input features(sepal_length, sepal_width, petal_length, petal_width)
3. Target variable: Species(Setosa, Versicolor, Virginica)
4. Number of classes: 3 species
   Each species has:
5. 50 samples

```python
# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load datasets
Iris = sns.load_dataset('iris')
# Display dataset structure
print("Shape of Dataset:")
print(Iris.shape)
print("\n Column Name:")
print(Iris.columns)
print("\n First five rows:")
print(Iris.head())
print("\n Basic Statistics:")
print(Iris.describe())
```

# Scatter plots
The following scatter plots show the relation between species and it is clearly visible that:
1. setosa is clearly seperated
2. other species overlap slightly
In pair plot setosa is clearly seperated in almost all plots with very distinct cluster and versicolor & verginica some overlap. It is harder to seperate using sepal features but petal length vs petal width shows perfect clustering.

```python
# sepal length vs sepal width
sns.scatterplot(
    x = 'sepal_length',
    y = 'sepal_width',
    hue = 'species',
    data = Iris
)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
#Petal Length vs Petal Width
sns.scatterplot(
    x = 'petal_length',
    y = 'petal_width',
    hue = 'species',
    data = Iris
)
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
# pair plot for feature importance intuition and clustering patterns
sns.pairplot(Iris, hue = 'species')
plt.show()
```

# Histogram
The following histogram show the visual analysis to understand the distribution of each feature according target variable(species) of the Iris dataset and as a result it has found that histplot for petal length and petal width is more informative. 

```python
import matplotlib.pyplot as plt

features = ["sepal_length", "sepal_width","petal_length","petal_width"]

plt.figure(figsize=(12, 10))

for i, col in enumerate(features):
    plt.subplot(2,2, i+1)
    sns.histplot(data = Iris, x = col, hue = "species", bins = 10, multiple = "stack")
    plt.title (f"Histogram of {col} by species")
plt.tight_layout()
plt.show()
```

# Box plot
The box plot for sepal length shows
1. some seperation but not perfect
2. Versicolor and Verginica overlap a lot
3. Setosa is slightly smaller but still close

```python
sns.boxplot(data = Iris, x = 'species', y = 'sepal_length', hue = 'species')
plt.title("Sepal Length Boxplot")
plt.show()
```

The boxplot for sepal length shows
1. Heavy overlap all three species
2. No clear seperation
3. High variability

```python
sns.boxplot(data = Iris, x = 'species', y = 'sepal_width', hue = 'species')
plt.title('Sepal Length Boxplot')
plt.show()
```

```python
sns.boxplot(data = Iris, x = 'species', y='petal_length', hue = 'species')
plt.title('Petal Length Boxplot')
plt.show()
```

The box plot for petal length shows:
1. setosa very small
2. versicolor medium
3. verginica large

Also shows clear seperation only some overlape between versicolor and verginica

```python
sns.boxplot(data = Iris, x = 'species', y='petal_width', hue = 'species')
plt.title('Petal Width Boxplot')
plt.show()
```

The box plot for petal width shows:
1. setosa very small
2. versicolor medium
3. verginica large

Also shows clear seperation only some overlape between versicolor and verginica
